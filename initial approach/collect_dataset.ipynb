```python
import os
from os import path

import shutil
```

```python
#collect train dataset
dataset_name = {
     '1' : 'realistic',
    # '2' : 'mid-journey',
     '3' : 'anime'
}

copy_pth = f'../contrastive_learning/data/train'

for dir_name in dataset_name.values():
    origin_pth = f'./{dir_name}'

    if not path.isdir(copy_pth+'/'+dir_name):
        os.makedirs(copy_pth+'/'+dir_name)

    dir_index_lst = os.listdir(origin_pth)

    index_lst = [file for file in dir_index_lst if not file.endswith(".npy")]
    
    for index in index_lst:
        if index == 'all_images':
            continue
        if int(index) > 300:
            continue
        else:
            img_pth_dir = origin_pth + '/' + index
            file_lst = os.listdir(img_pth_dir)
    
            for file in file_lst:
                if not path.exists(copy_pth+'/'+dir_name+'/'+file):
                    shutil.copy(origin_pth+'/'+index+'/'+file, copy_pth+'/'+dir_name+'/'+file)
```

```python
#collect test dataset
dataset_name = {
     '1' : 'realistic',
    # '2' : 'mid-journey',
     '3' : 'anime'
}

copy_pth = f'../contrastive_learning/data/test'

for dir_name in dataset_name.values():
    origin_pth = f'./{dir_name}'

    if not path.isdir(copy_pth+'/'+dir_name):
        os.makedirs(copy_pth+'/'+dir_name)

    dir_index_lst = os.listdir(origin_pth)
    index_lst = [file for file in dir_index_lst if not file.endswith(".npy")]
    
    for index in index_lst:
        if index == 'all_images':
            continue
        if int(index) <= 300:
            continue
        elif int(index) > 330:
            continue
        else:
            img_pth_dir = origin_pth + '/' + index
            file_lst = os.listdir(img_pth_dir)
    
            for file in file_lst:
                if not path.exists(copy_pth+'/'+dir_name+'/'+file):
                    shutil.copy(origin_pth+'/'+index+'/'+file, copy_pth+'/'+dir_name+'/'+file)
```
